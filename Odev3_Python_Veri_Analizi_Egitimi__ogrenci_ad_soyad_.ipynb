### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
import yfinance as yf
euro_21 = yf.download("EURTRY=X", start = "2021-01-01", end = "2021-12-31")

euro_21
```

2. Veriyi özet metodları ile inceleyiniz

```python
euro_21.head()
```

```python
euro_21.describe()
```

```python
euro_21.info()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
euro_21 = euro_21[["Adj Close"]]
euro_21.columns = ["Kapanış (TRY)"]
euro_21
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
euro_21["Normalize Kapanış (TRY)"] = euro_21["Kapanış (TRY)"] / euro_21.iloc[0,0]
euro_21
```

```python
bist30_21 = yf.download('XU030.IS', start = '2021-01-01', end = '2022-01-01')
bist30_21 = bist30_21[['Adj Close']]
bist30_21.columns = ['BIST30']
bist30_21['Normalize BIST30'] = bist30_21['BIST30'] / bist30_21.iloc[0, 0]
bist30_21
```

5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
import plotly.express as px
import plotly.graph_objects as go

bist30euro21= bist30_21.merge(euro_21, left_index = True, right_index = True)

bist30euro21['Normalize Kapanış (TRY)'] = bist30euro21['Kapanış (TRY)'] / bist30euro21.iloc[0, 2]
bist30euro21

```

```python
px.line(bist30euro21, y = ['Normalize BIST30', 'Normalize Kapanış (TRY)'],
        title='Normalize BIST 30 ve Normalize Kur (2021)',
        labels = dict({'Date':'Tarih', 'value': 'Kapanış (TRY)'}),
        )
```

```python

fig = go.Figure()
fig.add_trace(go.Bar(x = bist30euro21.index, y = bist30euro21['Normalize BIST30'], name = 'Normalize BIST30'))
fig.add_trace(go.Scatter(x = bist30euro21.index, y = bist30euro21['Normalize Kapanış (TRY)'], name = 'Normalize EUR/TRY'))

```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
import pandas as pd
indirme_listesi = ['XU030.IS', 'EURTRY=X']
bist30euro22 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2022-01-01', end = '2023-01-01')
  data = data['Adj Close'].rename(i)
  bist30euro22 = pd.concat([bist30euro22, data], axis = 1)

bist30euro22
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
bist30euro22.columns = ['BIST30', 'Kapanış (TRY)']
bist30euro22['Normalize BIST30'] = bist30euro22['BIST30'] / bist30euro21.iloc[0, 0] 
bist30euro22['Normalize Kapanış (TRY)'] = bist30euro22['Kapanış (TRY)'] / bist30euro21.iloc[0, 2]  
bist30euro22 = bist30euro22[bist30euro22['BIST30'] > 0]
bist30euro22
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
bist30euro = pd.concat([bist30euro21, bist30euro22], axis = 0)
bist30euro = bist30euro[bist30euro['BIST30'] > 0]
bist30euro
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
fig1 = px.line(bist30euro, y = ['Normalize BIST30', 'Normalize Kapanış (TRY)'])
fig1.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = bist30euro.index, y = bist30euro['Normalize BIST30'], name = 'Normalize BIST 30'))
fig.add_trace(go.Scatter(x = bist30euro.index, y = bist30euro['Normalize Kapanış (TRY)'], name = 'Normalize EUR/TRY'))

fig.update_layout(title = '2021-2022 Normalize BIST 30 ve Normalize Kur',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2)


fig.add_trace(
    go.Bar(x=bist30euro.index, y=bist30euro['Normalize Kapanış (TRY)'], name='Normalize EUR/TRY'),
    row=1, col=1
)


fig.add_trace(
    go.Bar(x=bist30euro.index, y=bist30euro['BIST30'], name='Normalize BIST30'),
    row=1, col=2
)

fig.update_layout(
    title='Normalize BIST 30 ve Normalize Kur (2021-2022)',
    xaxis_title='Tarih',
    yaxis_title='Kapanış (TRY)',
    hovermode='x',
    legend=dict(
        title='değerler',
    ),
    height=500,
    width=1000
)

fig.show()

```
